```python
import ltspice
import matplotlib.pyplot as plt
import numpy as np
```

```python
l = ltspice.Ltspice('Draft1_ohneCE.raw')
l.parse()
```

```python
f = l.getFrequency()
V_collector = l.getData('V(kollektor1)')
V_inp = l.getData('V(input)')
V_poti = l.getData('V(n003)')
V_Last = l.getData('V(n006)')
```

```python
plt.figure(figsize = (10,5))
plt.semilogx(f, np.log10(np.abs(V_collector))*20, label = 'Kollektorspannung')
plt.semilogx(f, 20*np.log10(np.abs(V_poti)), label = 'Potentiometerspannung')
plt.semilogx(f, 20*np.log10(np.abs(V_Last)), label = 'Lastspannung')
plt.semilogx(f, 20*np.log10(np.abs(V_inp)), label = 'Basispannung')
plt.ylabel('Verstärkung in dB')
plt.xlabel('Frequenz in Hz')
plt.grid(True)
plt.legend()
plt.title(r'Amplitudengang des Verstärkers ohne $C_{E}$')
plt.savefig('Verstärkungen_ohneCE.pdf', bbox_inches = 'tight')
plt.show()
```

```python
l = ltspice.Ltspice('Draft1_05V.raw')
l.parse()
```

```python
t = l.getTime()
V_base = l.getData('V(input)')
V_emitter = l.getData('V(n004)')
V_collector = l.getData('V(Kollektor1)')
V_out = l.getData('V(n003)')
```

```python
plt.figure(figsize = (10,5))
plt.plot(t, V_base-V_emitter)
plt.xlim([.8,.801])
plt.ylim([.4,.7])
plt.ylabel('Spannung in V')
plt.xlabel('Zeit in Sekunden')
plt.grid(True)
plt.title(r'Verlauf der Basis-Emitter-Spannung')
plt.savefig('BaseEmitterVoltage.pdf', bbox_inches = 'tight')
plt.show()
```

```python
plt.figure(figsize = (10,5))
plt.plot(t, V_out)
plt.xlim([.8,.801])
plt.ylim([0,9])
plt.ylabel('Spannung in V')
plt.xlabel('Zeit in Sekunden')
plt.grid(True)
plt.title(r'Verlauf der Ausgangsspannung (Poti)')
plt.savefig('outputvoltage.pdf', bbox_inches = 'tight')
plt.show()
```

```python
plt.figure(figsize = (10,5))
plt.plot(t, V_collector-V_emitter)
plt.xlim([.8,.801])
plt.ylim([0,7])
plt.ylabel('Spannung in V')
plt.xlabel('Zeit in Sekunden')
plt.grid(True)
plt.title(r'Verlauf der Kollektor-Emitter-Spannung')
plt.savefig('CollectorEmitterVoltage.pdf', bbox_inches = 'tight')
plt.show()
```
