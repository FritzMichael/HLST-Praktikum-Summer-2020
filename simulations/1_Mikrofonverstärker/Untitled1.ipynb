```python
import ltspice
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.rcParams.update({
    "text.usetex": True,
    "font.family": "serif",
    "font.sans-serif": ["Helvetica"]})
```

```python
l = ltspice.Ltspice('Mikrophonverst.raw')
l.parse()
```

```python
f = l.getTime()
noiseAmp = l.getData('V(onoise)')
```

```python
plt.figure(figsize = (10,5))
plt.semilogx(f, noiseAmp)
plt.ylabel(r'Rauschspannungsdichte in $\frac{\mathrm{V}}{\sqrt{\mathrm{Hz}}}$')
plt.xlabel('Frequenz in Hz')
plt.grid(True, which='both')
plt.xlim([10e-4, 10e7])
plt.xticks([1e-3,1e-2,1e-1,1e-0,1e1,1e2,1e3,1e4,1e5,1e6,1e7,1e8])
plt.title(r'Rauschverhalten am Ausgang')
plt.savefig('Rauschleistung.pdf', bbox_inches = 'tight')
plt.show()
```
