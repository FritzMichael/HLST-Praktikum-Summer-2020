```python
import ltspice
import matplotlib.pyplot as plt
import numpy as np
```

```python
l = ltspice.Ltspice('MillerOP_neu.raw')
l.parse()
```

```python
t = l.getTime()
V_out = l.getData('V(n008)')
```

```python
plt.figure(figsize = (10,5))
plt.plot(t, V_out)
plt.xlim([0.1,0.15])
plt.ylabel('Spannung in V')
plt.xlabel('Zeit in Sekunden')
plt.grid(True)
plt.title(r'Verlauf der Ausgangsspannung bei Brummspannung auf der Versorgung')
plt.savefig('Brumm.pdf', bbox_inches = 'tight')
plt.show()
```
